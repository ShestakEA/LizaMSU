```python
import matplotlib.pyplot as plt
import random as rand
import numpy as np
def rng(m=2**32, a=1103515245, c=12345):
    rng.current = (a * rng.current + c) % m
    return rng.current / m
rng.current = 1
random=[rng() for i in range(1000)]
random1=[(-1)*np.log(1-i) for i in random]
random2=[rand.expovariate(1) for i in range(1000)]

plt.figure(figsize=(13,4))
plt.subplot(1,2,1).hist(random1,bins=20)
plt.title("inverse transform method")
plt.xlim(left=0, right=10)

plt.subplot(1,2,2).hist(random2,bins=20)
plt.title("Истинная плотность распределения")
plt.xlim(left=0, right=10)

plt.show()
```

```python
from scipy import stats
N=1000
i=0
x=np.empty((0,1))
y=np.empty((0,1))
z=np.linspace(-4,4,100)
while i<N:
    x0=rand.uniform(-4,4)
    y0=rand.uniform(0,0.35)
    if y0<1/(np.pi*(1+x0**2)):
        x=np.append(x,x0)
        y=np.append(y,y0)
        i+=1
plt.figure(figsize=(8,6))
plt.plot(z,stats.cauchy().pdf(z),c="blue",linewidth=3,label="Истинная плотность")
plt.scatter(x,y,c="magenta",label="Попавшие точки")
plt.legend(loc="upper right")
plt.show()
```

```python
from matplotlib.patches import Circle
N=500
i=0
x=np.empty((0,1))
y=np.empty((0,1))

while i<N:
    x0=rand.uniform(-1,1)
    y0=rand.uniform(-1,1)
    if x0**2+y0**2<=1:
        x=np.append(x,x0)
        y=np.append(y,y0)
        i+=1
plt.figure(figsize=(8,6))
plt.scatter(x,y,c="magenta")
t = np.linspace(0, 2 * np.pi, 100)
plt.plot(np.sin(t), np.cos(t),linewidth=3)
plt.axis('equal')
plt.show()
```

```python
#n = 1000
random=np.array([rand.uniform(-1,1) for i in range(1000)])
x=np.array(random[1:])
y=np.array(random[:-1])
mask=x**2+y**2<=1
x=x[mask]
y=y[mask]
r=np.array(x**2+y**2)
x0 = x * np.sqrt(-2 * np.log(r) / r)
y0 = y * np.sqrt(-2 * np.log(r) / r)

fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].hist2d(x0, y0, bins=30)
ax[0].axis('equal')
ax[1].hist(x0, bins=30)
ax[2].hist(y0, bins=30)
ax[0].set_title("2d histogram")
ax[1].set_title("Histogram of x0")
ax[2].set_title("Histogram of y0")
plt.show()
```
