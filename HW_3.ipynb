```python
#задача №1
import numpy as np
import matplotlib.pyplot as plt
c=np.loadtxt("http://scipy-lectures.org/_downloads/populations.txt")
a=np.array(c.flatten()[::4])
plot1=plt.plot(a,np.array(c.flatten()[1::4]),"r--",a,np.array(c.flatten()[2::4]),"m",a,np.array(c.flatten()[3::4]),"g")
plt.xlabel("период времени")
plt.ylabel("популяция")
plt.legend(["зайцы","рыси","морковка"])
plt.show()
         
```

```python
#задача №4
c1=a[np.array(c.flatten()[1::4]).argmax()]
c2=a[np.array(c.flatten()[2::4]).argmax()]
c3=a[np.array(c.flatten()[3::4]).argmax()]
plt.plot(a,np.array(c.flatten()[1::4]),"r--",a,np.array(c.flatten()[2::4]),"m",a,np.array(c.flatten()[3::4]),"g")
plt.scatter([c1,c2,c3],[np.array(c.flatten()[1::4]).max(),np.array(c.flatten()[2::4]).max(),np.array(c.flatten()[3::4]).max()])
plt.legend(["зайцы","рыси","морковка","максимальная популция"])
```

```python
#задача №2
a1=(np.array(c.flatten()[1::4])*100)/np.array(c.flatten()[1::4]).sum()
a2=(np.array(c.flatten()[2::4])*100)/np.array(c.flatten()[2::4]).sum()
plt.bar(a,a1)
plt.bar(a,a2)
plt.legend(["зайцы","рыси"])
plt.show()
```

```python
#задача №3
print("Mean values:", np.mean(c[:,1:4], axis = 0))
print("Standard deviation values:", np.std(c[:,1:4], axis = 0))

plt.figure(figsize=(7, 15))
plt.subplot(3,1,1)
plt.hist(c[:,1], bins=10)
plt.axvline(c[:,1].mean(), color = 'red')
plt.title("зайцы")

plt.subplot(3,1,2)
plt.hist(c[:,2], bins=6, label = "рыси")
plt.axvline(c[:,2].mean(), color = 'red')
plt.title("рыси")

plt.subplot(3,1,3)
plt.hist(c[:,3], bins=9)
plt.axvline(c[:,3].mean(), color = 'red')
plt.title("морковка")

plt.show()
```

```python
#задача №6
plt.figure(figsize=(10, 7))
plt.plot(c[:,0], c[:,1])
plt.plot(c[:,0], c[:,2])
plt.plot(c[:,0], c[:,3])
plt.axhline(50000, color = "grey")
plt.scatter(c[:,0][c[:,1] > 50000], c[:,1][c[:,1] > 50000], 50, c='red')
plt.scatter(c[:,0][c[:,2] > 50000], c[:,2][c[:,2] > 50000], 50, c='red')
plt.xticks(np.arange(c[:,0].min(), c[:,0].max(), 3))

plt.xlabel("Время")
plt.ylabel("Популяция")
plt.legend(["зайцы","рыси","морковка","5000",">5000"])
plt.show()
```

```python
#задача №7
print("2 года с наименьшей популяцией зайцев")
print(c[np.argmin(c[:,1]),0])
mask = c[:,1] != np.min(c[:,1])
print(c[:,0][mask][np.argmin(c[:,1][mask])])

print("\n2 года с нименьшей популяцией рысей")
print(c[np.argmin(c[:,2]),0])
mask = c[:,2] != np.min(c[:,2])
print(c[:,0][mask][np.argmin(c[:,2][mask])])

print("\n2 года с наименьшим урожаем моркови")
print(c[np.argmin(c[:,3]),0])
mask = c[:,3] != np.min(c[:,3])
print(c[:,0][mask][np.argmin(c[:,3][mask])])
```
