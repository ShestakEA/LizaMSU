```python
from scipy.stats import binom
import numpy as np

M = 10
y = np.array([
    7, 7, 1, 3, 2, 1, 5, 1, 4, 0, 
    2, 3, 3, 1, 2, 2, 3, 1, 4, 0, 
    8, 3, 5, 3, 0, 0, 7, 1, 1, 3, 
    1, 3, 2, 4, 1, 6, 2, 2, 4, 1, 
    3, 1, 1, 2, 7, 3, 3, 2, 2, 2
])
```

```python
np.random.seed(1)
N=len(y.flatten())
pi = np.array([0.5, 0.5])
mu = np.random.rand(2)
eps = 1e-6
```

```python
num = 0

while True:
    print(f'iteraton #{num}:')
    print(f'pi = {pi}')
    print(f'mu = {mu}')
    print()
    
    num += 1
    
    #E-step
    #rik = p(xi = k|yi,θ(t−1)) = p(yi|xi = k, μ(t−1)k)π(t−1)k/Summ(s,p(yi|xi = s, μ(t−1)s )π(t−1))
    r = binom.pmf(y[:, None], M, mu[None, :]) * pi[None, :]
    r /= np.sum(r, axis=1)[:, None]
    
    #M-step
    mu0=mu.copy()
    pi0=pi.copy()
    
    #π(t)k =Summ(i,rik)/N
    pi=r.sum(axis=0)/N
    
    #μ(t)k =Summ(i,yi*rik)/Summ(i,rik)*M
    mu = np.sum(y[:, None] * r, axis=0) / r.sum(axis=0) / M

    
    if np.abs(pi - pi0).sum() < eps and np.abs(mu - mu0).sum() < eps:
        break
    
```

```python
print('pi =',pi," ",'mu =', mu)
print('число итераций =',num)
```
